#**Data Visualization with Matplotlib**


###**Histograms**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to create histograms for specified numerical columns/attributes
def plot_numerical_histograms(dataframe, numerical_features):
    num_cols = len(numerical_features)  # Number of numerical columns
    num_rows = (num_cols + 1) // 2  # Number of rows in the subplot grid

    fig, axes = plt.subplots(num_rows, 2, figsize=(12, 5*num_rows))

    # Flatten axes if there's only one row
    if num_rows == 1:
        axes = axes.reshape(1, -1)

    # Plot histograms for each numerical column
    for i, column in enumerate(numerical_features):
        row = i // 2
        col = i % 2
        ax = axes[row, col]
        dataframe[column].plot(kind='hist', ax=ax)
        ax.set_title(f'Histogram of {column}')

        skewness = dataframe[column].skew()
        ax.text(0.95, 0.85, f'Skewness: {skewness:.2f}', transform=ax.transAxes, ha='right', va='top', bbox=dict(facecolor='white', alpha=0.5))

        # Identify outliers
        q1 = dataframe[column].quantile(0.25)
        q3 = dataframe[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)][column]
        if not outliers.empty:
            ax.text(0.95, 0.75, f'Outliers: {outliers.values}', transform=ax.transAxes, ha='right', va='top', bbox=dict(facecolor='white', alpha=0.5))

    # If number of columns is odd, remove the empty subplot
    if num_cols % 2 != 0:
        fig.delaxes(axes[-1, -1])

    plt.tight_layout()
    plt.show()

# Define the numerical features to plot histograms for
numerical_features = ["BMI", "PhysicalHealth", "MentalHealth", "SleepTime"]

# Call the function with your DataFrame and numerical features
plot_numerical_histograms(dataframe, numerical_features)
```

###Note:
The correlation matrix analysis has revealed that no attribute pair demonstrated a correlation coefficient approaching 1, signifying a lack of strong linear association between attributes. Therefore, applying the Univariate Feature Selction Algorithm to individually assess each attribute's predictive capability and selecting k=10 best features.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif  # for classification tasks
import pandas as pd

# uni variate feature selection

# Separate features (X) and target variable (y)
X = dataframe.drop('HeartDisease', axis=1)  # Adjust 'target_column' with the name of your target column
y = dataframe['HeartDisease']

# Initialize SelectKBest with the desired number of features and the scoring function
# Here, we'll use f_classif for classification tasks. For regression, use f_regression.
k_best = SelectKBest(score_func=f_classif, k=10)  # Adjust 'k' with the desired number of features

# Fit SelectKBest to your data
X_selected = k_best.fit_transform(X, y)

# Get the indices of the selected features
selected_indices = k_best.get_support(indices=True)

# Get the names of the selected features
selected_features = X.columns[selected_indices]

# Print the selected features
print("Selected features:", selected_features)
```

```python
selected_features = ['Smoking', 'Stroke', 'PhysicalHealth', 'DiffWalking', 'Sex',
                     'AgeCategory', 'Diabetic', 'PhysicalActivity', 'KidneyDisease',
                     'SkinCancer']

# Keep only the selected features in the DataFrame
data_selected = dataframe[selected_features]
print(data_selected)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Assuming X contains your features and y is your target variable
X = data_selected
y = dataframe['HeartDisease']
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree classifier
dt_classifier = DecisionTreeClassifier()

# Train the model
dt_classifier.fit(X_train, y_train)

# Make predictions
y_pred = dt_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
